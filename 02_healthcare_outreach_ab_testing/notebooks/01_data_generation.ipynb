# 01 Data Generation

This notebook creates a synthetic healthcare outreach A/B test dataset. The business question is whether a new outreach message increases Annual Wellness Visit completion compared with a standard message.

The dataset is fully synthetic. It does not contain real patient data, PHI, or clinical claims.


## Setup

We import the helper functions from `src/data_generation.py`. The notebook is designed to run from the `notebooks/` folder, so we add the project root to `sys.path`.


```python
from pathlib import Path
import sys

import pandas as pd
import matplotlib.pyplot as plt

PROJECT_ROOT = Path.cwd().resolve().parents[0]
sys.path.append(str(PROJECT_ROOT / "src"))

from data_generation import generate_outreach_data, save_outreach_data
```

## Generate Synthetic Members

Treatment assignment is randomized. The treatment group has a modest positive effect, so completion rates should be near 24% for control and 29% for treatment.


```python
df = generate_outreach_data(n_members=8000, random_state=42)
df.head()
```

## Basic Data Checks

These checks confirm row count, columns, treatment assignment, and approximate completion rates.


```python
print("Shape:", df.shape)
print("\nExperiment group counts:")
print(df["experiment_group"].value_counts())
print("\nCompletion rate by group:")
print(df.groupby("experiment_group")["awv_completed"].mean())
```

## Save Raw and Processed Files

The raw file contains the core experiment data. The processed file adds simple flags that are useful for segment analysis later.


```python
raw_path, processed_path = save_outreach_data(df, PROJECT_ROOT)
print("Raw data saved to:", raw_path)
print("Processed data saved to:", processed_path)
```
