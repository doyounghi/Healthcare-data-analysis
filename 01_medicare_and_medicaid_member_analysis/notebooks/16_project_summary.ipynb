# Project Overview

This project is a synthetic healthcare analytics and machine learning portfolio project.

The goal is to practice end-to-end data science workflow using a healthcare-style dataset, including:

- synthetic data generation
- exploratory data analysis
- feature engineering
- classification modeling
- regression modeling
- model interpretation
- model comparison
- leakage prevention

Because the dataset is synthetic, all findings should be interpreted as simulated modeling practice rather than real-world healthcare evidence.

## 1. Dataset summary

The dataset represents a synthetic member-level healthcare population.

It includes simulated features related to:

- demographics
- region
- plan type
- chronic condition burden
- engagement score
- SDOH risk score
- PCP attribution
- prior AWV history
- ED visits
- inpatient admissions
- monthly cost
- AWV completion

## 3. Model Classification

### 3a. AWV Completion Classification

The classification task focused on predicting whether a member completed an Annual Wellness Visit.

The target variable was:

`awv_completed`

To prevent leakage, the following columns were excluded from the classification models:

- `member_id`
- `awv_completed`
- `monthly_cost`
- `high_cost_member`

`monthly_cost` and `high_cost_member` were excluded because they are cost-related outcome variables and should not be used to predict AWV completion.

Logistic Regression and Decision Tree Classifier were compared using ROC AUC, F1 score, precision, recall, and confusion matrix results.

The Logistic Regression model was selected as the stronger baseline classification model because it achieved better ROC AUC and F1 score than the Decision Tree Classifier.

#### Clean Modeling Feature Set

Across the modeling notebooks, the cleaned predictor set was kept consistent to reduce leakage risk and improve comparability.

For AWV classification and monthly cost regression, the core raw predictors were:

- `age`
- `gender`
- `region`
- `plan_type`
- `sdoh_risk_score`
- `chronic_condition_count`
- `engagement_score`
- `pcp_attributed_24mo`
- `prior_awv_rate`
- `ed_visits`
- `ip_admits`

Grouped features, full-dataset quartile features, identifiers, target-derived fields, and redundant engineered variables were excluded from modeling when they created leakage risk or made interpretation less clean.

### 3b. Monthly Cost Regression

The regression task focused on predicting simulated monthly healthcare cost.

The target variable was:

`monthly_cost`

To prevent leakage, the following columns were excluded from the regression models:

- `member_id`
- `monthly_cost`
- `high_cost_member`
- `awv_completed`

`high_cost_member` was excluded because it is derived from `monthly_cost` and would leak target information.

`awv_completed` was excluded to keep the baseline cost prediction task focused on member characteristics, risk, access, and utilization variables.

```text
Gradient Boosting Regressor

This model was selected based primarily on test RMSE.

RMSE was prioritized because monthly healthcare cost prediction can be strongly affected by large prediction errors. MAE and R² were also reviewed as supporting metrics.

The tuned Gradient Boosting model was not selected because tuning did not materially improve performance over the baseline Gradient Boosting model.



## 7. Interpretation caution

Model results should be interpreted carefully.

For classification:

- Logistic Regression coefficients are conditional associations, not causal effects.
- Odds ratios describe changes in predicted odds, not guaranteed real-world impact.
- Classification performance depends on the selected threshold.

For regression:

- Regression coefficients are conditional associations, not causal effects.
- Standardized coefficients help compare numeric predictors but do not eliminate multicollinearity.
- Tree-based feature importance shows predictive usefulness inside the fitted model, not causal importance.
- Synthetic data results reflect the assumptions built into the data-generation process.

## 8. Limitations

1. The dataset is synthetic and does not represent real member outcomes.
2. Results depend on the data-generation assumptions.
3. The final models are only best among the models and hyperparameter grids tested.
4. Feature importance and coefficients should not be interpreted causally.
5. A model with stronger test performance may not always be the best operational model if interpretability, stability, or deployment simplicity matters more.

## 9. Portfolio Takeaway

This project demonstrates an end-to-end healthcare analytics and machine learning workflow.

The project includes synthetic data generation, EDA, feature engineering, leakage review, classification modeling, regression modeling, coefficient interpretation, threshold tuning, residual diagnostics, tree-based models, ensemble models, hyperparameter tuning, and final model comparison.

A key lesson from the project is that stronger models are not always more complex. Regularization and hyperparameter tuning did not always produce meaningful improvements. Gradient Boosting performed best in this synthetic setup, while simpler linear models remained competitive and more interpretable.

The final results should be interpreted as portfolio-level modeling practice using synthetic data, not as real-world healthcare evidence.
